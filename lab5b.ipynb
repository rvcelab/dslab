```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as mp
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data=pd.read_excel("C:/Users/DS1/Downloads/random_forest_dataset (1).xlsx")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.head(5)
```

```python
data.isna().sum()
```

```python
data['R1']=data['R1'].fillna(data['R1'].mean())
```

```python
data['T1']=data['T1'].fillna(data['T1'].mean())
```

```python
data['Grade']=data['Grade'].fillna(data['Grade'].mode())
data['Grade'].mode()
```

```python
data.isna().sum()
```

```python
x=data.drop(['Sl No','Grade','USN','Name','Title'],axis=1)
x.isnull().sum()
```

```python
y=data['Grade']
y.head
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
y_train.isna().sum()
```

```python
rfc=RandomForestClassifier(n_estimators=10,random_state=42)
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
